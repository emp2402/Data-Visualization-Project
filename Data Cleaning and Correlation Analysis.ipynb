```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import mean
```

```python
#open the file
with open("diabetes.csv") as f:
    df = pd.read_csv(f)

    #generate clean data set
    cleandf = df.copy()

    #remove rows having "smoking" or "drinking" attributes outside their established ranges
    for x in cleandf.index:
        if (cleandf.loc[x, "smoking"] or cleandf.loc[x, "drinking"]) not in [1,2,3]:
            cleandf.drop(x, inplace = True)

    #Save the clean dataset
    cleandf.to_csv("cleaned_data.csv", index=False)
```

```python
#Correlation matrix
with open("cleaned_data.csv") as f:
    df = pd.read_csv(f)
    corr_matrix = df.corr()
    diabetes_corr = corr_matrix["Diabetes"].sort_values()
    print(diabetes_corr)
```
